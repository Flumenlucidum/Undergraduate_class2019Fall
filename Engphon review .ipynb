```python
#generate pulse train   samples, note conversion to float32 array

F0=100; Fend=int(sr/2); s=np.zeros(len(t))
for freq in range(F0,Fend+1,F0):
    theta =t*2*np.pi*freq
    tmp=amp*np.sin(theta)
    s=s+tmp
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(t[0:200],s[0:200])
ax.set_xlabel('time(msec)')  # 이 그림은 waveform
```

```python
def hz2w(F,sr):
    NyFreq=sr/2
    w=F/NyFreq*np.pi
    return w 

def resonance(srate,F,BW):
    a2=np.exp(-hz2w(BW,srate))
    omega=F*2*np.pi/srate
    a1=-2*np.sqrt(a2)*np.cos(omega)
    a=np.array([1,a1,a2])
    b=np.array([sum(a)])
    return a,b
```

```python
RG=0 #freq of glottal resonator 산맥의 위치 
BWG=100 #bandwith 산맥의 퍼진 정도 
a,b =resonance(sr,RG,BWG)
```

```python
s=lfilter(b,a,s,axis=0)
```

```python
ipd.Audio(s,rate=sr)
```

```python
nFFT=nSamp
amp=[];
for n in range(0,nFFT):
    omega=2*np.pi*n/nFFT #angular velocity
    z=np.exp(omega*1j)**(np.arange(0,nSamp))
# 2 pi *0....nsamp/nsamp-> 0부터 2pi 까지 가는 벡터
#한바퀴를 도는 셈 이게 첫루프 -- 이후 nSamp바퀴까지 
#(e ^wi)^[0....nsamp]   **는 몇 승  z가  nsamp-1개 존재
#이것을 s와 inner product
    amp.append(np.abs(np.dot(s,z))) 
    #abs 가 절댓값 #amp의 길이는 nsamp와 같아 amp에는 허수 안들어가 있어
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
freq=np.arange(1,nFFT+1)*sr/nFFT;
ax.plot(freq,amp)   # freq 100... 10000
ax.set_xlabel('frequency(Hz)')
ax.set_ylabel('amplitude')  #막대의 개수는 bar의 개수와 같아 
```
